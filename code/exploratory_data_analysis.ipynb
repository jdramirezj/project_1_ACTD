#### Import required packages 

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Import data

#### Create feature of percentage (%) of approved curricular units by semester

```python
data['perc_approved_sem1'] = data['Curricular units 1st sem (approved)']/data['Curricular units 1st sem (enrolled)']
data['perc_approved_sem2'] = data['Curricular units 2nd sem (approved)']/data['Curricular units 2nd sem (enrolled)']
```

```python
data['perc_approved_sem2'].describe()
```

```python
sns.boxplot(data=data, x='Target', y='Age at enrollment')
```

```python
sns.boxplot(data=data, x='Target', y='perc_approved_sem2')
```

```python
sns.boxplot(data=data, x='Debtor', y='perc_approved_sem1')
```

```python
sns.boxplot(data=data, x='Scholarship holder', y='perc_approved_sem1')
```

```python
data['scholarship_debtor'] = (data['Scholarship holder'] & data['Debtor']).astype(int)
```

```python
sns.boxplot(data=data, x='scholarship_debtor', y='perc_approved_sem1')
```
