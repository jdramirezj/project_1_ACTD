#### Import required packages 

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()
```

#### Import data

```python
data = pd.read_csv('../data/data.csv', sep=';')
```

#### Show the head of the data

```python
data.head()
```

#### Show the data columns

```python
data.columns
```

#### Create feature of percentage (%) of approved curricular units by semester

```python
data['perc_approved_sem1'] = data['Curricular units 1st sem (approved)']/data['Curricular units 1st sem (enrolled)']
data['perc_approved_sem2'] = data['Curricular units 2nd sem (approved)']/data['Curricular units 2nd sem (enrolled)']
```

#### Change the target variable for it being one when the student dropped his career

```python
data['Target'] = np.where(data['Target'] == 'Dropout', 1, 0)
```

#### Check if the new features are between 0 and 1

```python
data['perc_approved_sem1'].describe()
```

```python
data['perc_approved_sem2'].describe()
```

#### Scatter plot of the relationship between the percentage of approved units in the 1st and 2nd semester

```python
plt.scatter(x=data['perc_approved_sem1']*100, y=data['perc_approved_sem2']*100)
plt.title('Comparison of Approval Percentages: Semester 1 vs. Semester 2', fontsize=16)
plt.xlabel('Percentage Approved - Semester 1 (%)', fontsize=14)
plt.ylabel('Percentage Approved - Semester 2 (%)', fontsize=14)
plt.show()
```

#### Check the relationship between the Age and the Target

```python
sns.boxplot(data=data, x='Target', y='Age at enrollment')
plt.title('Age by Target', fontsize=16)
plt.xlabel('Target', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.show()
```

#### Check the relationship between the percentage of unit approved in the second semester and the Target

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=data['Target'], y=data['perc_approved_sem2']*100)
plt.title('Percentage of units approved in the Second Semester by Target (%)', fontsize=16)
plt.xlabel('Target', fontsize=14)
plt.ylabel('Percentage of units approved in the Second Semester (%)', fontsize=14)
plt.show()
```

#### Check the relationship between the percentage of unit approved in the first semester and Debtor

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='Debtor', y='perc_approved_sem1')
plt.title('Percentage of units approved in the First Semester by Debtor (%)', fontsize=16)
plt.xlabel('Debtor', fontsize=14)
plt.ylabel('Percentage of units approved in the First Semester (%)', fontsize=14)
plt.show()
```

#### Check the relationship between the percentage of unit approved in the first semester and Scholarship holder

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='Scholarship holder', y='perc_approved_sem1')
plt.title('Percentage of units approved in the First Semester by Scholarship Holder (%)', fontsize=16)
plt.xlabel('Scholarship Holder', fontsize=14)
plt.ylabel('Percentage of units approved in the First Semester (%)', fontsize=14)
plt.show()
```

```python
sns.regplot(x=data["Mother's qualification"], y=data['perc_approved_sem1'], color='red')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='Target', y="Mother's qualification")
plt.title("Mother's qualification by Scholarship Holder", fontsize=16)
plt.xlabel('Scholarship Holder', fontsize=14)
plt.ylabel("Mother's qualification", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='Target', y="Father's qualification")
plt.title("Father's qualification by Scholarship Holder", fontsize=16)
plt.xlabel('Scholarship Holder', fontsize=14)
plt.ylabel("Father's qualification", fontsize=14)
plt.show()
```
